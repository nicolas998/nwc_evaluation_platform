```python
mask = np.load('../data/mask_1D_ix.npz', allow_pickle = True)
hrr_p = np.load('../data/hrrr_pixel_area.npy')
```

```python
def usgs_get_area(usgs_id):
    pos = np.where(mask['usgs_id'] == usgs_id)[0]
    mm = mask['indexes1d'][pos]
    return hrr_p[mm[0][0]].sum()/1e6
```

```python
m = netCDF4.Dataset('../data/2019_nwc_error.nc')
```

```python
f = open('../data/hrrr_usgs_areas.csv','w')
f.write('gauge, area\n')
for usgs_id in mask['usgs_id']:
    f.write('%s, %.2f\n' % (usgs_id, usgs_get_area(usgs_id)))
f.close()    
    
```
